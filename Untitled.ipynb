```python
import os
import pandas as pd
import pyprind
```

Download URL: "http://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"

```python
basepath='data/aclImdb'
```

```python
labels = {'pos': 1, 'neg': 0}
```

```python
pbad = pyprind.ProgBar(50000)
```

```python
df = pd.DataFrame()
```

```python
for s in ('test', 'train'):
    for l in ('pos', 'neg'):
        
        path = os.path.join(basepath, s, l)
        
        for file in os.listdir(path):
            with open(os.path.join(path, file),
                         'r', encoding='utf-8') as infile:
                
                txt = infile.read()
                
            df = df.append([[txt, labels[l]]],
                              ignore_index=True)
            
            pbar.update()
```

```python
df.columns = ['review', 'sentiment']
```
